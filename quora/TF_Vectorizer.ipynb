```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

docs=['I really like to play football and watch football on TV. Football is my favourite sport',
      'My favourite TV serial is really awesome and I enjoy watching it',
      'I dont really want to talk about my favourite colour',
      'My favourite subject in school is computer science, because computer science is really awesome.'
     ]
```

```python
cv = CountVectorizer(lowercase=True, analyzer='word', stop_words='english')
tf_idf = TfidfTransformer()
word_count_matrix = cv.fit_transform(docs)

tf_idf.fit(word_count_matrix) # Tfidf uses word counts matrix to compute the final score

df_idf = pd.DataFrame(tf_idf.idf_, index=cv.get_feature_names(), columns=['idf_scores'])
df_idf.sort_values(by=['idf_scores'])
```

```python
tf_idf_matrix = tf_idf.transform(word_count_matrix).toarray()
first_doc_scores = tf_idf_matrix[0]

print(tf_idf_matrix)
print(first_doc_scores)

df_tf_idf = pd.DataFrame(first_doc_scores, index=cv.get_feature_names(), columns=['tf_idf_scores'])
df_tf_idf.sort_values(by=['tf_idf_scores'])

```
